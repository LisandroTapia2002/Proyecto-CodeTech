```python
### Plan de Ahorro ###

import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

# Inicializar los ingresos totales y el historial de transacciones
ingresos_totales = 0
historial_transacciones = []
balance = {'Gastos Fijos': 0, 'Ahorro': 0, 'Ocio': 0}

def agregar_ingreso():
    """Agrega un ingreso total al balance."""
    global ingresos_totales
    try:
        monto = float(input("Ingresa el monto total de ingresos: "))
        if monto <= 0:
            raise ValueError("El monto debe ser positivo.")
        ingresos_totales += monto
        titulo = input("Título del ingreso: ").strip()
        fecha = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        historial_transacciones.append({'Fecha': fecha, 'Monto': monto, 'Tipo': 'Ingreso', 'Título': titulo})
        print(f"Ingreso de {monto} añadido correctamente.")
    except ValueError as e:
        print(f"Error: {e}. Intenta nuevamente.")

def seleccionar_plan_ahorro():
    """Selecciona un plan de ahorro y devuelve las proporciones correspondientes."""
    while True:
        print("\nSelecciona un plan de ahorro:")
        print("1. 60% Gastos Fijos, 30% Ahorro, 10% Ocio")
        print("2. 50% Gastos Fijos, 35% Ahorro, 15% Ocio")
        print("3. 70% Gastos Fijos, 20% Ahorro, 10% Ocio")

        try:
            opcion = int(input("Elige una opción (1, 2, o 3): "))
            if opcion == 1:
                return 0.60, 0.30, 0.10
            elif opcion == 2:
                return 0.50, 0.35, 0.15
            elif opcion == 3:
                return 0.70, 0.20, 0.10
            else:
                print("Opción no válida, selecciona nuevamente.")
        except ValueError:
            print("Entrada no válida, ingresa un número (1, 2, o 3).")

def calcular_distribucion(ingresos, plan):
    """Calcula la distribución del ingreso según el plan de ahorro seleccionado."""
    gastos_fijos = ingresos * plan[0]
    ahorro = ingresos * plan[1]
    ocio = ingresos * plan[2]
    return gastos_fijos, ahorro, ocio

def mostrar_grafica(gastos_fijos, ahorro, ocio):
    """Muestra una gráfica de barras con la distribución del ingreso."""
    categorias = ['Gastos Fijos', 'Ahorro', 'Ocio']
    valores = [gastos_fijos, ahorro, ocio]

    plt.figure(figsize=(10, 5))
    barras = plt.bar(categorias, valores, color=['blue', 'green', 'red'])
    plt.xlabel('Categorías')
    plt.ylabel('Monto')
    plt.title('Distribución del Ingreso')

    # Añadir etiquetas con la cantidad exacta de dinero
    for barra in barras:
        altura = barra.get_height()
        plt.annotate(f'{altura:.2f}', xy=(barra.get_x() + barra.get_width() / 2, altura),
                     xytext=(0, 3),  # 3 puntos de desplazamiento vertical
                     textcoords="offset points", ha='center', va='bottom')
    plt.show()

def actualizar_balance(tipo, monto):
    """Actualiza el balance de cada categoría."""
    if tipo in balance:
        balance[tipo] += monto

def mostrar_balance():
    """Muestra una gráfica de barras con el balance actual."""
    categorias = list(balance.keys())
    valores = list(balance.values())

    plt.figure(figsize=(10, 5))
    barras = plt.bar(categorias, valores, color=['blue', 'green', 'red'])
    plt.xlabel('Categorías')
    plt.ylabel('Monto')
    plt.title('Balance Actual')

    # Añadir etiquetas con la cantidad exacta de dinero
    for barra in barras:
        altura = barra.get_height()
        plt.annotate(f'{altura:.2f}', xy=(barra.get_x() + barra.get_width() / 2, altura),
                     xytext=(0, 3),  # 3 puntos de desplazamiento vertical
                     textcoords="offset points", ha='center', va='bottom')
    plt.show()

def agregar_transaccion():
    """Agrega una transacción al balance."""
    while True:
        print("\nSelecciona el tipo de transacción:")
        print("1. Gastos Fijos")
        print("2. Ahorro")
        print("3. Ocio")

        try:
            opcion = int(input("Elige una opción (1, 2, o 3): "))
            if opcion == 1:
                tipo = "Gastos Fijos"
            elif opcion == 2:
                tipo = "Ahorro"
            elif opcion == 3:
                tipo = "Ocio"
            else:
                print("Opción no válida, selecciona nuevamente.")
                continue

            monto = float(input(f"Ingresa el monto para {tipo}: "))
            if monto <= 0:
                raise ValueError("El monto debe ser positivo.")
            titulo = input("Título de la transacción: ").strip()
            fecha = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
            historial_transacciones.append({'Fecha': fecha, 'Monto': monto, 'Tipo': tipo, 'Título': titulo})
            actualizar_balance(tipo, monto)
            print(f"Transacción de {monto} para {tipo} añadida correctamente.")
            mostrar_balance()
            break
        except ValueError as e:
            print(f"Error: {e}. Intenta nuevamente.")

def mostrar_historial():
    """Muestra el historial de transacciones."""
    if historial_transacciones:
        df = pd.DataFrame(historial_transacciones)
        print("\nHistorial de Transacciones:")
        print(df)
    else:
        print("\nNo hay transacciones registradas.")

def menu():
    """Muestra el menú principal y maneja la interacción del usuario."""
    while True:
        print("\n" + "="*50)
        print("                   MENÚ PRINCIPAL")
        print("="*50)
        print("1. Agregar ingreso total")
        print("2. Seleccionar plan de ahorro")
        print("3. Agregar transacción")
        print("4. Mostrar historial de transacciones")
        print("5. Salir")
        print("="*50)

        try:
            opcion = int(input("Elige una opción: "))
            
            if opcion == 1:
                agregar_ingreso()
            elif opcion == 2:
                if ingresos_totales > 0:
                    gastos_fijos, ahorro, ocio = calcular_distribucion(ingresos_totales, seleccionar_plan_ahorro())
                    mostrar_grafica(gastos_fijos, ahorro, ocio)
                else:
                    print("Por favor, ingrese un monto total antes de seleccionar un plan de ahorro.")
            elif opcion == 3:
                agregar_transaccion()
            elif opcion == 4:
                mostrar_historial()
            elif opcion == 5:
                print("\n¡Hasta luego! Gracias por usar Cachichen.")
                break
            else:
                print("Opción no válida. Intenta nuevamente.")
        except ValueError:
            print("Entrada no válida, ingresa un número del 1 al 5.")

menu()
```
